```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset["classification_yes"].value_counts()
```

```python
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dep=dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(indep, dep, test_size=1/3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
dataset
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

param_grid={'kernel':['linear', 'rbf', 'sigmoid'],
           'gamma':['auto','scale'],'C':[10,100,1000,2000,3000]}

grid=GridSearchCV(SVC(probability=True),param_grid, refit=True, verbose=3, n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
```

```python
re=grid.cv_results_
grid_predictions=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix #procedure for confusion matrix
cm=confusion_matrix(y_test, grid_predictions)
```

```python
from sklearn.metrics import classification_report #procedure for classification report
clf_report=classification_report(y_test, grid_predictions)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions, average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
print("The Confusion Matrix:\n",cm)
```

```python
print("The report:\n",clf_report)
```

```python
from sklearn.metrics import roc_auc_score
print("Roc_auc_score is:",roc_auc_score(y_test,grid.predict_proba(X_test)[:,1]))
```
